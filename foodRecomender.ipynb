```python
foods = foods[['nom', 'description', 'type', 'difficulte', 'calories','ing',
               'propriete', 'ingredient','cout_total', 'image', 'rating']]
```

```python
# Lets create a pivot table

food_pivot = foods.pivot_table(index=['propriete', 'nom','type'], values=['rating'])
```

```python
# Now import our clustering algoritm which is Nearest Neighbors this is an unsupervised ml algo
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(food_sparse)
```

```python
def recommend_food(food_name):
    if food_name not in food_pivot.index:
        print(f"'{food_name}' not found in the recommendation system.")
        return

    food_id = np.where(food_pivot.index == food_name)[0]
    distance, suggestion = model.kneighbors(food_pivot.iloc[food_id,:].values.reshape(1,-1), n_neighbors=6 )

    for i in range(len(suggestion)):
        foods = food_pivot.index[suggestion[i]]
        for j in foods:
            if j == food_name:
                print(f"You searched '{food_name}'\n")
                print("The suggested foods are:\n")
            else:
                print(j)
```
